```python
## one more data is there in zomato file country-code
```

```python
data2=pd.read_excel("country-code.xlsx")
```

```python
data2.head()
```

```python
# in first data set also have country code variable and 
#in 2nd data set also have country code variable let's 
#merge this 2 data set.
```

```python
final_data=pd.merge(data1,data2, on="Country Code", how="left")
```

```python
final_data.head()
```

```python
# counting the country basis on obs
final_data.Country.value_counts()
```

```python
# Here i got all the contry name 
country_name=final_data.Country.value_counts().index
```

```python
country_name
```

```python
# Here i got all the values of country
country_val=final_data.Country.value_counts().values
```

```python
country_val
```

```python
# Now trying to plot aa pie chart to see the distribution of zomato over the world
plt.pie(country_val[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
#Observation: From above pie i understood that zomato
#services are maximum in india after that united 
#states and then united kingdom
```

```python
final_data.columns
```

```python
# Now i'm grouping aggregate rating rating colour and rating text 
```

```python
# Here i used size() options to display the grouping result
# and i used reset_index option to get the obs and the data in proper format
# and then  used renmae option to change the number of rating previously it was 0
Rating=final_data.groupby(["Aggregate rating",'Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating number"})
```

```python
Rating
```

```python
#Observation: From above data i understand that-
#Most of the people did not rate the app (Not rated)
#The rating between 1.8 to 2.4 is ---> poor
#The rating between 2.5 to 3.4 are ---> Average
#The rating between 3.5 to 3.9 is ---> Good
#The rating between 4.0 to 4.4 are ---> Very good
#The rating between 4.5 to 4.9 is ---> Excellent
```

```python
# Now i'm going to do a bar plot against ratings and number of ratings
# and here i used same matplotlib syntax like previous one to increase 
#the sixe of graph for better viualization
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x="Aggregate rating",y="Rating number",data=Rating)
```

```python
# Now i'm going to do a bar plot of same variables like previous one 
#but this time i'm gonna add hue to differntiate the exact colour which
#is mentioned in dataset
#And this plot i used the colour according to the dataset
sns.barplot(x="Aggregate rating",y="Rating number",hue="Rating color",data=Rating,palette=["white","red","orange","yellow","green","green"])
```

```python
#Observation: From the above graph i understand that 
#maximum number of rating between 2.5 to 3.4 //check 
#the colour that orange colour is maximum in the graph
```
